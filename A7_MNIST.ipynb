```python
import tensorflow as tf;  #importing tensorflow library
import numpy as np        #importing numpy library for numerical operations
import matplotlib.pyplot as plt  #importing matplot library for ploting
import random             #importing random for defining random values
```

```python
mnist = tf.keras.datasets.mnist; #load the data from keras
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()  #loaded data is stored into test and training sets
print(x_train.shape)               #training set consists of 60000 images of size 28x28 pixel size
print(y_train.shape)             #no. of images label
print(x_test.shape)             #test set consists of 10000 images of size 28x28 pixel size
print(y_test.shape)              #no. of images label
```

```python
x_train = tf.keras.utils.normalize(x_train) #normalise the training set
x_test = tf.keras.utils.normalize(x_test)  #normalise the test set
```

```python
#Reshape all images into single rows
X = x_train.reshape(60000,784) # convert 2D(28x28) image into a single 1D(784) image vector
y = y_train;  
```

```python
a= X.shape   #image count is stored in variable a
print(a)     #print the value of count
input_layer_size  = a[1];  # 28x28 pixel Input Image of Digits
hidden_layer_size = 25;   # 25 hidden units
num_labels = 10;          # 10 labels, from 0 to 9
lambd = 2;                #initializing constant lambda
alpha = 2;                #initializing constant alpha
```

```python
#initializing the values of theta
theta1 = np.zeros([input_layer_size+1, hidden_layer_size]);
theta2 = np.zeros([hidden_layer_size+1, num_labels]);
```

```python
#updating the values of theta
epsilon_init = 0.1;
theta1 = np.random.uniform(low=0, high=1, size=(input_layer_size+1, hidden_layer_size)) * 2 * epsilon_init - epsilon_init;

epsilon_init = 0.1;
theta2 = np.random.uniform(low=0, high=1, size=(hidden_layer_size+1, num_labels)) * 2 * epsilon_init - epsilon_init;
```

```python
#defining sigmoid function
def sigmoid(x):
  return 1/(1 + np.exp(-x)) 
  
```

```python
#defining sigmoid gradient function
def sigmoidGradient(z):
  s  = 1/(1+np.exp(-z));
  return s*(1-s);
```

```python
#element wise multiplying of theta and training set X
Z1 = np.dot(np.concatenate((np.ones([a[0],1]), X), axis = 1),theta1);
```

```python
first_Layer_Output = sigmoid(Z1); #first layer output
```

```python
a= first_Layer_Output.shape  #shape of the first layer
Z2 = np.dot(np.concatenate((np.ones([a[0],1]), first_Layer_Output), axis = 1),theta2); #estimated output of layer 2
second_Layer_Output = sigmoid(Z2);        #getting output of second layer
second_Layer_Output.shape                 #shape of second layer
```

```python
YMat = np.zeros(second_Layer_Output.shape); #creating a matrix of same dimension as that of second layer
a = second_Layer_Output.shape;              #storing its shape 
YMat.shape 
```

```python
for i in range(0,a[0]):
    YMat[i, y[i]] = 1;
```

```python
no_of_Interations = 100;  #take no. of iterations = 100
J = np.zeros([no_of_Interations,1]); #create a column vector of 100 rows 

for i in range(0,no_of_Interations):
  a= X.shape;               #shape of the training set is stored in variable a
  Z1 = np.dot(np.concatenate((np.ones([a[0],1]), X), axis = 1),theta1);   #Output is calculated as element wise multiplication
  first_Layer_Output = sigmoid(Z1);                                 #finding sigmoid of it i.e probabilistic value between 0-1

  a= first_Layer_Output.shape #shape of the output of first layer i.e sigmoid is stored
  Z2 = np.dot(np.concatenate((np.ones([a[0],1]), first_Layer_Output), axis = 1),theta2);  #Output is calculated by element wise multiplication
  second_Layer_Output = sigmoid(Z2);                                #finding sigmoid of output i.e probabilistic value between 0-1
  
  J[i] = (-1/a[1])*np.sum(YMat*np.log(second_Layer_Output) + (1-YMat)*np.log(1-second_Layer_Output)) + (lambd/(2*a[1]))*np.sum(theta1[1:,:]*theta1[1:,:]) + (lambd/(2*a[1]))*np.sum(theta2[1:,:]*theta2[1:,:]); #finding the cost function value

  da2dz2 = (second_Layer_Output - YMat);                          #error is calculated as difference of (predicted value and actual value)
  da1dz1 = da2dz2.dot(np.transpose(theta2[1:,:]))*sigmoidGradient(Z1);
  
  b = second_Layer_Output.shape;
  Theta2_grad = np.dot(np.transpose(np.concatenate((np.ones([a[0],1]), first_Layer_Output),1)),da2dz2);     #theta2_grad = A1 * d(J)/dZ[2]
  Theta2_grad = Theta2_grad +lambd*np.concatenate((np.zeros([1, b[1]]), theta2[1:,:]),0);                   #regularized theta2
  
  Theta1_grad = np.dot(np.transpose(np.concatenate((np.ones([a[0],1]), X),1)),da1dz1);                      #theta1_grad = A1 * d(J)/dZ[1]
  Theta1_grad = Theta1_grad + lambd*np.concatenate((np.zeros([1, a[1]]), theta1[1:,:]),0);                  #regularized theta1
  
  Theta2_grad = Theta2_grad/a[0];                   #averaging is done by total number of elements present
  Theta1_grad = Theta1_grad/a[0];                   #averaging is done by total number of elements present
  
  theta1 = theta1- alpha*Theta1_grad; #updating theta after every iteration
  theta2 = theta2 - alpha*Theta2_grad; #updating theta after every iteration

  if(i%10 == 0):                 #plotting for every 10 iterations
    plt.plot(J[0:i]);            #figure no.
    plt.title('Cost function reduction for lambda = '+ str(lambd)+ ' and alpha = '+ str(alpha)); #  title of the figure
    plt.xlabel('Iteration number'); #label of figure x
    plt.ylabel('Cost function (J\theta)');#label of figure y
    plt.show();          #show figure
    plt.pause(0.05);     # Pause for 0.05 seconds
```

```python
#Evaluation of training set%%
a= X.shape;
Z1 = np.dot(np.concatenate((np.ones([a[0],1]), X), axis = 1),theta1);  #calculating preactivation Z1 =X^T*theta1 #^T indicates  transpose
first_Layer_Output = sigmoid(Z1);  #calcluating the activation value using sigmoid function for Z1 so as to limit the value between 0-1

a= first_Layer_Output.shape
Z2 = np.dot(np.concatenate((np.ones([a[0],1]), first_Layer_Output), axis = 1),theta2);  #calculating preactivation Z2 =X^T*theta2 #^T indicates transpose 
second_Layer_Output = sigmoid(Z2);  #calcluating the activation value using sigmoid function for Z2 so as to limit the value between 0-1

I = np.zeros(y.shape);           #initialising matrix of shape of y 
training_Error_Percentage = 0;   #initialising value
for i in range(0,a[0]):          
    I[i] = np.argmax(second_Layer_Output[i,:]); #returns the maximum value along the axis
    if(I[i] == y[i]):
      training_Error_Percentage = training_Error_Percentage+1;   #accuracy calculation
training_Error_Percentage = 100*training_Error_Percentage/a[0];
print(training_Error_Percentage)                                 #percentage of accuracy
```

```python
#Evaluation of test set
X_test = x_test.reshape(10000,784);

a= X_test.shape;
Z1 = np.dot(np.concatenate((np.ones([a[0],1]), X_test), axis = 1),theta1); #calculating preactivation Z1 =X^T*theta1 #^T indicates  transpose
first_Layer_Output = sigmoid(Z1);  #calcluating the activation value using sigmoid function for Z1 so as to limit the value between 0-1

a= first_Layer_Output.shape
Z2 = np.dot(np.concatenate((np.ones([a[0],1]), first_Layer_Output), axis = 1),theta2);#calculating preactivation Z2 =X^T*theta2 #^T indicates transpose 
second_Layer_Output = sigmoid(Z2);          #calcluating the activation value using sigmoid function for Z2 so as to limit the value between 0-1

I = np.zeros(y_test.shape);      #initialising matrix of shape of y 
test_Error_Percentage = 0;       #initialising value
for i in range(0,a[0]):
    I[i] = np.argmax(second_Layer_Output[i,:]);    #returns the maximum value along the axis
    if(I[i] == y_test[i]):
      test_Error_Percentage = test_Error_Percentage+1;    #accuracy calculation
test_Error_Percentage = 100*test_Error_Percentage/a[0];
print(test_Error_Percentage)                              #percentage of accuracy
```
